```python
import pandas as pd
import requests
import json
import numpy as np
```

```python
url = 'https://graphql.anilist.co'
```

```python
# Here we define our query as a multi-line string
query = '''
    query Query {
  MediaListCollection(userName: "leolion023", type: ANIME, status: COMPLETED) {
    lists {
      name
      entries {
        mediaId
        media {
          title {
            english
            romaji
          }
          format
        }
      }
    }
  }
}
'''

response = requests.post(url, json={'query': query}) # Make the HTTP Api request
```

```python
json = response.json() #convert response to json
json.keys() #find keys
```

```python
df = pd.json_normalize(json['data'])
df
```

```python
df_1 = pd.json_normalize(df['MediaListCollection.lists'])
df_1
```

```python
df_2 =pd.json_normalize(df_1[0])
df_2
```

```python
df_3 = pd.json_normalize(df_2['entries']).T
df_3.head()
```

```python
df_4 = pd.json_normalize(df_3[0])
df_4
```

```python
df_5 = df_4.loc[df_4['media.format']=='TV']
```
